# Learn Linear Functions 线性函数

This unit will show you how to develop linear equations from two points.

线性函数的斜率：

$M = \frac{y2 - y1}{x2 - x1}$

线性函数的截距：

$y = Mx + b$

```python
x1 = 1
y1 = 3
x2 = 4
y2 = 9

# 斜率 m
m = (y2 - y1) / (x2 - x1)

# 纵截距 b
b = y1 - m*x1

print(f'The equation is y = {m}*x + {b}')
```

```python
import matplotlib.pyplot as plt

"""计算线性函数表达式"""
x1 = 1
y1 = 3
x2 = 4
y2 = 9

# 斜率 m
m = (y2 - y1) / (x2 - x1)
# 纵截距 b
b = y1 - m*x1

print(f'函数表达式 y = {m}*x + {b}')

"""set some varieties to express the range of x-axis and y-axis"""
x_min = -10
x_max = 10
y_min = -10
y_max = 10

"""matplotlib"""
fig, ax = plt.subplots()

# 设置坐标轴范围
plt.axis([x_min, x_max, y_min, y_max])

# 横线和竖线
plt.plot([0, 0], [y_min, y_max], 'b')
plt.plot([x_min, x_max], [0, 0], 'b')

# 计算图像边界的点的坐标
y3 = m*x_min + b
y4 = m*x_max + b

# 根据图像边界点绘制直线，这样可以使线条在整个图像中显示
plt.plot([x_min, x_max], [y3, y4], 'r')

plt.show()
```

## use code to solve math problems 使用代码解决数学问题

the popuparity of a city is 55,000 in 2010, and 76,000 in 2012. Predict the popularity in 2016.

```python
import matplotlib.pyplot as plt
import numpy as np

"""计算函数表达式"""
x1 = 0 # use 0 to express 2010
y1 = 55 # use 55 to express 55,000
x2 = 2
y2 = 76

# 斜率和纵截距
m = (y2 - y1) / (x2 - x1)
b = y1 - m*x1

print(f'y = {m}*x + {b}')

"""matplotlib"""
x_min = 0
x_max = 10
y_min = 40
y_max = 150

fig, ax = plt.subplots()
plt.axis([x_min, x_max, y_min, y_max])

# 坐标轴标题，图像标题，网格，刻度线间隔
ax.set_xlabel('years')
ax.set_ylabel('popularity')
ax.set_title('the popularity of a city')
ax.grid(True)
ax.set_xticks(np.arange(x_min, x_max, 1))
ax.set_yticks(np.arange(y_min, y_max, 10))

# 计算边界处的坐标
y3 = m*x_min + b
y4 = m*x_max +b

# 绘制线
plt.plot([x_min, x_max], [y3, y4], 'r')

plt.show()
```
