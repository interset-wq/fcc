# Learn How to Graph Systems of Equations 在一个坐标系中绘制多个图像

This unit will focus on graphing two equations on the same coordinate plane, so that you can see where the lines intersect.

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10

# 通过matplotlib绘制的实际上是折线图
# 如果是曲线图就需要通过更多的点才能使曲线看起来平滑
points = 10 * (xmax - xmin)
x = np.linspace(xmin, xmax, points)

# 引入中文字体，使matplotlib可以显示中文
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots()

plt.axis([xmin, xmax, ymin, ymax])
plt.title('多个图像')
# ax.set_title('two line') # 作用和plt.title()相同，用于设置图像标题
plt.plot([xmin, xmax], [0, 0], 'b')
plt.plot([0, 0], [ymin, ymax], 'b')
ax.set_xlabel('x')
ax.set_ylabel('y')

# line 1
y = 2 * x
plt.plot(x, y, 'y')

# line 2
y = x**2 - 3
plt.plot(x, y, 'g')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xmin = -10
xmax = 10
ymin = -10
ymax = 10

# 足够多的点，使曲线平滑
points = 10 * (xmax - xmin)
x = np.linspace(xmin, xmax, points)

# 中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots()

plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], 'k')
plt.plot([0, 0], [ymin, ymax], 'k')
# 坐标轴和图像标题
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('更多曲线')
# 坐标轴刻度间隔
ax.set_xticks(np.arange(xmin, xmax, 1))
ax.set_yticks(np.arange(ymin, ymax, 1))

# 如果某个区域有多种颜色，后面的会覆盖前面的

# line 1
y1 = x + 6
plt.plot(x, y1, 'k', label='y=x+6')
plt.fill_between(x, y1, ymax, facecolor='r')

# line 2
y2 = x + 3
plt.plot(x, y2, 'k')
plt.fill_between(x, y2, y1, facecolor='y')

# line 3
y3 = x - 1
plt.plot(x, y3, 'k')
plt.fill_between(x, y3, y2, facecolor='g')

# line 4
y4 = x - 4
plt.plot(x, y4, 'k')
plt.fill_between(x, y4, y3, facecolor='b')

# plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
# 通过滚动条修改函数方程
from ipywidgets import interactive

def f(m: float, b: float, zoom: float) -> None:
    """通过斜率和截距绘制直线，通过zoom控制坐标轴范围"""

    xmin = -zoom
    xmax = zoom
    ymin = -zoom
    ymax = zoom

    points = 10 * (xmax - ymin)
    x = np.linspace(xmin, xmax, points)

    plt.axis([xmin, xmax, ymin, ymax])
    plt.plot([xmin, xmax], [0, 0], 'k')
    plt.plot([0, 0], [ymin, ymax], 'k')
    plt.title('y = mx + b')

    plt.plot(x, m*x + b)
    plt.show()

# 使用滚动条绘制不同的图像
interactive_plot = interactive(f, m=(-9, 9), b=(-9, 9), zoom=(1, 100))
# 相当于调用函数
interactive_plot
```
